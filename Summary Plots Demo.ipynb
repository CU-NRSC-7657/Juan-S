## Individual Components

**Average Waveform**

This function was adopted from a earlier shared function used by the lab which I hardly tweaked. The output is a mean waveform from the neuron across all channels of the neuropixels probe.

```python
probe_folder = glob(path+'\*imec0*')[0] #Change the probe number as necessary
bin_file = glob(probe_folder+'\*ap.bin')[0] #Change the probe number as necessary
imec_meta = sglx.readMeta(probe_folder+'\\')
sampRate = float(imec_meta['imSampRate']) #get sampling rate (Hz)

mean_wf = sq.mean_waveform(rawdata=bin_file,times=units_df.times[15],channels=385,sampling_rate = float(sampRate))
```

```python
for i in mean_wf:
    plt.plot(i)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    plt.xticks([])
```

**Spatial Receptive Fields**

`rf_array` generates spatial receptive field arrays based on spike triggered averages with different tau values from stimulus frames in the random noise matrix. `rf_grid` takes that array data and displays it as a grid of heatmaps(?)

```python
test_rf_array = summary.rf_array(spike_data = units_df, probe='imec0', unit=15, stim_data = matrix_data['luminance_data'], 
                 stim_times = matrix_data['luminance_times'],num_taus=10)
```

```python
test_rf_grid = summary.rf_grid(data=test_rf_array, colormap ='viridis',interp='none',title='Luminance',nrows=5,ncols=2)
```

**Gratings**

Both `psth_line_overlay` and `psth_bars` take spike times and stimulus times to display an interpretation of peristimulus time histograms across different stimulus conditions. `psth_bars` is (in my opinion) prettier and more legible, but has limitations. `psth_line_overlay` gives a better idea of baseline firing rate and variance, but (in my opinion) is much more difficult to interpret.

```python
gratings_orientation = gratings_data[gratings_data['condition']=='orientation']
```

```python
test_ori = summary.psth_line_overlay(spike_data=units_df, probe='imec0', unit=15, stim_data=gratings_orientation, 
                                     condition='ori',title='Orientation', pre=0.5, post=2.5,binsize=0.05,variance=True)
```

```python
test_bars = summary.psth_bars(spike_data=units_df, probe='imec0', unit=15, stim_data=gratings_orientation, 
                                     condition='ori',title='Orientation', pre=0.5, post=2.5,binsize=0.05)
```
